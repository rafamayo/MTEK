```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
# import control as ctrl
```

```python
# Parameter des Systems
m = 1  # Masse (analog zur Trägheit des Blutes)
k = 1  # Federkonstante (elastische Eigenschaften der Blutgefäße)
b = 0.5  # Dämpfung (Widerstand der Blutgefäße)
```

```python
# Übertragungsfunktion H(s) = 1 / (ms^2 + bs + k)
numerator = [1]  # Zähler der Übertragungsfunktion
denominator = [m, b, k]  # Nenner der Übertragungsfunktion
```

```python
# Zeitverhalten simulieren (Impulsantwort)
system = signal.TransferFunction(numerator, denominator)
t, y_impulse = signal.impulse(system)

```

```python
# Frequenzverhalten simulieren (Bode-Diagramm)
w, mag, phase = signal.bode(system)
```

```python
# Plot: Zeitverhalten (Impulsantwort)
plt.figure()
plt.plot(t, y_impulse)
plt.title('Zeitverhalten: Impulsantwort')
plt.xlabel('Zeit [s]')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
# Plot: Frequenzverhalten (Bode-Diagramm)
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(w, mag)  # Magnitude plot
plt.title('Bode-Diagramm')
plt.ylabel('Magnitude [dB]')
plt.grid(True)
plt.subplot(2, 1, 2)
plt.semilogx(w, phase)  # Phase plot
plt.xlabel('Frequenz [rad/s]')
plt.ylabel('Phase [degrees]')
plt.grid(True)
```

```python
# Zeige alle Plots an
plt.tight_layout()
plt.show()
```
